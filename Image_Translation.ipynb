## Translating or shifting the position of an image using OpenCV

```python
# Importing required libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading and displaying an input image

plt.rcParams["figure.figsize"] = (10,10)

image = cv2.imread("images/input.jpg")
plt.imshow(image)
```

```python
# Converting image to RGB channel

rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
# Getting the size of image

print(rgb_image.shape)
```

```python
# Getting Height & width of image in separate variable

Height, Width = rgb_image.shape[:2]
```

```python
# Moving / transalting image by distance, hence calaulating the distance first
# X distance = image height / 4.0
# Y distance = image width / 4.0

X_dist = Width / 4.0
Y_dist = Height / 4.0
```

```python
# Translation Matrix

#     | 1  0  Tx |
# T = | 0  1  Ty |

# Creating the translation matrix

Translation_Matrix = np.float32([[1, 0, X_dist], [0, 1, Y_dist]])
```

```python
# We will use warpAffine to translate the image using Translation_Matrix

Translated_image = cv2.warpAffine(rgb_image, Translation_Matrix, (Width, Height))
```

```python
# Displaying the transformed image

plt.imshow(Translated_image)
```
