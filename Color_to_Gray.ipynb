## Converting a Color image to a Gray-scale image using OpenCV

```python
# Importing Libraries

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2

%matplotlib inline
```

```python
# Displaying a color image

plt.rcParams["figure.figsize"] = (10,10)

color_img = img.imread('images/Color_to_Gray.jpg')
plt.imshow(color_img)
```

```python
print(type(color_img))
```

```python
print('Image Shape', color_img.shape)
```

```python
# Coverting image to Gray_scale image

gray_image = cv2.cvtColor(color_img, cv2.COLOR_RGB2GRAY)
```

```python
# Displaying Gray Image

plt.imshow(gray_image, cmap='gray')
```
