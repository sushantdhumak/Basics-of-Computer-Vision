## Image scaling & re-sizing using OpenCV

```python
# Importing required libraries

import cv2
import matplotlib.pyplot as plt
```

```python
# Reading and displaying an input image

plt.rcParams["figure.figsize"] = (10,10)

image = cv2.imread("images/input.jpg")
plt.imshow(image)
```

```python
# Converting image to RGB channel

rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
# Scaling image by making image size to half

Scaled_image1 = cv2.resize(rgb_image, None, fx=0.5, fy=0.5)
plt.imshow(Scaled_image1)
```

```python
# Re-sizing image by making image size to (900, 400) using interpolation

Scaled_image2 = cv2.resize(rgb_image, (900, 900), interpolation = cv2.INTER_AREA)
plt.imshow(Scaled_image2)
```
