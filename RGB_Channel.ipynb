## Separating RGB channels of a color image using OpenCV

```python
# Importing Libraries

import matplotlib.pyplot as plt
import matplotlib.image as img

%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (10,10)

color_image = img.imread('images/RGB_Channel.jpg')
plt.imshow(color_image)
```

```python
r = color_image[:,:,0]
g = color_image[:,:,1]
b = color_image[:,:,2]

f, (x1, x2, x3) = plt.subplots(1, 3, figsize = (20,10))

x1.set_title('R Channel')
x1.imshow(r, cmap='gray')
x2.set_title('G Channel')
x2.imshow(g, cmap='gray')
x3.set_title('B Channel')
x3.imshow(b, cmap='gray')
```
