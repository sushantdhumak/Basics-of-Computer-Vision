### Basics of Computer Vision

## Changing green background of an image using OpenCV

```python
# Importing required libraries

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2

%matplotlib inline
```

```python
# Displaying input image
plt.rcParams["figure.figsize"] = (9, 5)

Original_img_b4_conv = cv2.imread('images/Green_Ironman.jpg')
plt.imshow(Original_img_b4_conv)
```

```python
# Converting image from BGR to RGB color format

Original_img = cv2.cvtColor(Original_img_b4_conv, cv2.COLOR_BGR2RGB)
plt.imshow(Original_img)
```

```python
# Getting shape/size of input image i.e. (Row, Column, Color Channels)

print('Shape of the image', Original_img.shape)
```

```python
# Defining color thershold for Green color

lower_green = np.array([0,150,0]) 
upper_green = np.array([132,255,132])
```

```python
# Creating and displaying the mask area

mask = cv2.inRange(Original_img, lower_green, upper_green)
print(type(mask))
plt.imshow(mask, cmap='gray')
```

```python
# Applying mask area to Original image

masked_img = np.copy(Original_img)
masked_img[mask != 0] = [0,0,0]

plt.imshow(masked_img)
```

```python
# Converting & displaying background image

back_img_b4_conv = cv2.imread('images/Background.jpg')
back_img = cv2.cvtColor(back_img_b4_conv, cv2.COLOR_BGR2RGB)
plt.imshow(back_img)
print('Image shape', back_img.shape)
```

```python
# Cropping the background to the size of original image

Cropped_back_img = back_img[0:720, 0:1280]
plt.imshow(Cropped_back_img)
```

```python
# Applying mask to the background image

Cropped_back_img[mask == 0] = [0,0,0]
plt.imshow(Cropped_back_img)
```

```python
# Merging masked and cropped image to get the final output image

Flying_Ironman_img = masked_img + Cropped_back_img
plt.imshow(Flying_Ironman_img)
```
