## Rotating an image around its center using OpenCV

```python
# Importing required libraries

import cv2
import matplotlib.pyplot as plt
```

```python
# Reading and displaying an input image

plt.rcParams["figure.figsize"] = (10,10)

image = cv2.imread("images/input.jpg")
plt.imshow(image)
```

```python
# Converting image to RGB channel

rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
# Getting the size of image

print(rgb_image.shape)
```

```python
# Getting Height & width of image in separate variable

Height, Width = rgb_image.shape[:2]
```

```python
# Getting the center of the image

CenterX = Width / 2.0
CenterY = Height / 2.0
```

```python
# Creating the rotation matrix

Rotation_Matrix = cv2.getRotationMatrix2D((CenterX, CenterY), 45, 1)
```

```python
# We will use warpAffine to rotate the image using Rotation_Matrix

Rotated_image = cv2.warpAffine(rgb_image, Rotation_Matrix, (Width, Height))
```

```python
# Displaying the transformed image

plt.imshow(Rotated_image)
```
